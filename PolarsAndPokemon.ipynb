## `.over()` **Window Functions** 
They're a little cleaner and you don't have to do as many mental gymnastics around group_by objects. # Putting it all together
After going through each step, I'd put all the operations together in one function so I don't have to worry about tracking intermediate variables or forgetting where my data was modified. Once I have this, I can format my data nicely in one fell swoop. This method of method-chaining was popularized by Matt Harrison with his Effective Pandas book/videos. It reads like a recipe and is contained in one function.

```python
pokemon_df.estimated_size('kb')
```

```python
import re

def tweak(df_: pl.DataFrame):
    '''cleans the data'''
    starters = ['Bulbasaur', 'Charmander', 'Squirtle', 'Chikorita', 'Cyndaquil', 'Totodile', 
                'Treecko', 'Torchic', 'Mudkip', 'Turtwig', 'Chimchar', 'Piplup', 
                'Snivy', 'Tepig', 'Oshawott', 'Chespin', 'Fennekin', 'Froakie', 
                'Rowlet', 'Litten', 'Popplio', 'Grookey', 'Scorbunny', 'Sobble']
    terms = ['the', 'z']
    return (df_
            .lazy()
            .with_columns(pl.col(['Type 1', 'Type 2']).cast(pl.Categorical),
                          pl.col(['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 
                                  'Sp. Def','Speed']).cast(pl.Int16),
                          pl.col('Generation').cast(pl.Int8),
                          pl.col('Name').str.replace('Mega', '-Mega')
                                        .str.replace('Primal', '-Primal')
                                        .str.replace('Origin', '-Origin'),
                          original = pl.when(pl.col("Generation") == 1).then(pl.lit('Original Gangsta')).otherwise(None),
                          starter = pl.when(pl.col("Name").is_in(starters)).then(pl.lit('Starter pokemon')).otherwise(None),
                          weird_name = pl.col('Name').str.contains('\d'),
                          others_of_same_types = pl.col('Name')
                                                   .unique()
                                                   .over(['Type 1', 'Type 2'], 
                                                         mapping_strategy='join')
                                                   .list.join(', '),
                          )
            .with_columns(contains_term = pl.col('others_of_same_types').map_elements(lambda s: next((t for t in terms if t in s), None)),
                          contains_terms = pl.col('others_of_same_types')
                                             .map_elements(lambda s: list(set(re.findall('|'.join(terms), s))), 
                                                           return_dtype=pl.List(pl.Utf8))
                                             .list.join(', '), #sets aren't supported in arrow data
                          num_contained_terms = pl.col('others_of_same_types').map_elements(lambda s: len(re.findall('|'.join(terms), s))),
                          )
            .unique()
           )

pokemon_df = tweak(pokemon_raw).collect()
pokemon_df
```

## Eager vs Lazy Evaluation
With the lazy API, Polars doesn't run each query line-by-line but instead processes the full query end-to-end. To get the most out of Polars it is important that you use the lazy API because:

* the lazy API allows Polars to apply automatic query optimization with the query optimizer
* the lazy API allows you to work with larger than memory datasets using streaming
* the lazy API can catch schema errors before processing the data

**Eager evaluation** is where code is evaluated as soon as you run the code. 
**Lazy evaluation** is where running a line of code means that the underlying logic is added to a query plan rather than being evaluated.

```python
scan_df.collect(streaming=True).write_parquet('larger_than_mem_df.parquet')
```

You can also make a loaded dataset use the Lazy API with `.lazy()`. This enables query optimization - instead of running things line by line, which is extremely helpful for large datasets.

## LazyFrame considerations:

several lessons learned:
* `.pipe()` apparently breaks streaming - while I was using .`pipe` to chain before, I noticed that it would cause errors for larger files
* `.inspect()` is fantastic for viewing your df at the desired intermediate steps. It can show exactly the transformations along the way and show exactly how much or little each step is doing

```python
import re
terms = ['a', 'b', 'c']

def print_and_return(ldf_):
    '''takes a lazy df and prints shape'''
    print(ldf_.shape)
    return ldf_

def scrub_a_dub(df_): 
    '''utilizes the .inspect() to see the df at inermediate steps'''
    return (df_.lazy()
            .inspect()                       #shows full shape and small sample
            .map_batches(lambda x: print_and_return(x), 
                         streamable=True)    #can use this to only show the shape - but inspect is better
            .with_columns(
               poke_per_group = pl.col('Name').unique().over(['Type 1', 'Type 2'], mapping_strategy='join').list.join(', '),
               count = pl.n_unique('Name').over(['Type 1', 'Type 2']),
               original = pl.when(pl.col("Generation") == 1).then(1).otherwise(-1).alias('Original'),
               has_term = pl.col('others_of_same_types').str.contains('|'.join(terms)),
               has_term2 = pl.col('others_of_same_types').map_elements(lambda s: next((t for t in terms if t in s), None)),
               each_matched_term = pl.col('others_of_same_types')
                                     .map_elements(lambda s: re.findall('|'.join(terms), s))
                                     .list.join(', '),
               num_matched_terms = pl.col('others_of_same_types')
                                     .map_elements(lambda s: len(re.findall('|'.join(terms), s)),
                                                  return_dtype=pl.Int64),
               matched_terms = pl.col('others_of_same_types')
                                 .map_elements(lambda s: list(re.findall('|'.join(terms), s)), 
                                               return_dtype=pl.List(pl.Utf8)), #sometimes will present problems if you don't explicitly define a return value
               )
            .collect(streaming=True)
            .write_parquet('pokemon_lazy.parquet')
           )

pokemon_raw.pipe(scrub_a_dub)
pl.scan_parquet('pokemon_lazy.parquet').fetch(5)
```
